## What is a Linear Regression?

A **linear regression model** predicts a continuous value by fitting a straight line through a set of observations.

## Single Input Variable

For a single input variable, the model has the form:

$$
\hat{y} = mx + b
$$

where:

* $x$ is the input variable,
* $\hat{y}$ is the predicted value,
* $m$ is the slope (gradient) of the line, and
* $b$ is the intercept.


### Example

For example below we have the observed heights (cm) and weights (kg) or a sample of 20 people. Which we show on a scatter plot.

```python
heights_cm = np.array([
    172, 158, 184, 166, 177,
    191, 163, 180, 155, 174,
    188, 169, 182, 160, 176,
    193, 167, 185, 171, 179
], dtype=float)

weights_kg = np.array([
    74, 61, 92, 68, 83,
    106, 57, 78, 52, 89,
    84, 98, 76, 64, 71,
    112, 73, 87, 69, 95
], dtype=float)

plot_scatter_plot(
    x=heights_cm, 
    y=weights_kg, 
    x_label="Height (cm)", 
    y_label="Weight (kg)"
)
```

An example of a linear regression model would be the following (note: this is not the best fit, this model has been guessed):

$$\hat{\operatorname{w}} = 0.9 \times \operatorname{h} - 80$$

where:

* $h$ (height) is the input variable,
* $\hat{w}$ (weight) is the predicted value,
* $0.9$ is the slope (gradient) of the line, and
* $-80$ is the intercept.

We can define this as a Python function and add to our plot:

```python
def linear_regression_single_input(height: np.array) -> np.array:
    return 0.9 * height - 80

# Heights covering the full plotted range
line_heights = np.linspace(
    heights_cm.min(),
    heights_cm.max(),
    100
)

# Weights predicted across our full plotted range of heights
line_weights = linear_regression_single_input(line_heights)

plot_scatter_plot(
    x=heights_cm, 
    y=weights_kg, 
    x_label="Height (cm)", 
    y_label="Weight (kg)",
    line_values=(line_heights, line_weights)
)
```

Although we cannot say anything about it's quality at this stage the above line can be used to predict the weight from a given height.

For example if a person is 175cm tall we can predict their weight (kg):

```python
height = 175
predicted_weight = linear_regression_single_input(height)

print("Using our linear regression we have:") 
print(f"Height (Observed): {height}cm")
print(f"Weight (Predicted): {predicted_weight}kg.")
```

This can also be vislualised as:

```python
plot_scatter_plot(
    x=heights_cm, 
    y=weights_kg, 
    x_label="Height (cm)", 
    y_label="Weight (kg)",
    line_values=(line_heights, line_weights),
    single_prediction = height
)
```

## Multiple Input Variables

Linear regressions are not just limited to one input variable. In general, a linear regression is a model that predicts a continuous value as a weighted sum of one or more input variables and an intercept.

For \(n\) input variables, a linear regression model has the form:

$$
\hat{y} = w_1x_1 + w_2x_2 + \cdots + w_nx_n + b
$$

where:

- $x_1, x_2, \ldots, x_n$ are the input variables,
- $w_1, w_2, \ldots, w_n$ are the weights associated with each input variable,
- $b$ is the intercept, and
- $\hat{y}$ is the predicted value.

### Example

For example below we have the observed heights (cm), age (years) and weights (kg) or a sample of 20 people. Which we show on a scatter plot.

```python
heights_cm = np.array([
    172, 158, 184, 166, 177,
    191, 163, 180, 155, 174,
    188, 169, 182, 160, 176,
    193, 167, 185, 171, 179
], dtype=float)

age_yrs = np.array([
    34, 52, 28, 41, 37,
    24, 59, 46, 31, 63,
    27, 55, 39, 22, 48,
    35, 44, 61, 29, 50
], dtype=float)

weights_kg = np.array([
    74, 61, 92, 68, 83,
    106, 57, 78, 52, 89,
    84, 98, 76, 64, 71,
    112, 73, 87, 69, 95
], dtype=float)

plot_scatter_plot(
    x=[heights_cm, age_yrs], 
    y=weights_kg, 
    x_label=["Height (cm)", "Age (years)"], 
    y_label="Weight (kg)"
)
```

An example of a linear regression model with two input variables would be the following (note: this is still a guessed model, not the best fit):

$$\hat{\operatorname{w}} = 0.9 \times \operatorname{h} - 0.2 \times \operatorname{a} - 70$$

where:

* $h$ (height) is the first input variable,
* $a$ (age) is the second input variable,
* $\hat{w}$ (weight) is the predicted value,
* $0.9$ and $-0.2$ are the weights associated with each input variable, and
* $-70$ is the intercept.

We can define this as a Python function and add the resulting plane to our plot:

```python
def linear_regression_multiple_inputs(height: float, age: float) -> float:
    return 0.9 * height - 0.2 * age - 70

# Heights and ages covering the full plotted range
surface_heights, surface_ages = np.meshgrid(
    np.linspace(heights_cm.min(), heights_cm.max(), 30),
    np.linspace(age_yrs.min(), age_yrs.max(), 30)
)

# Weights predicted across our full plotted range of heights and ages
surface_weights = linear_regression_multiple_inputs(surface_heights, surface_ages)

plot_scatter_plot(
    x=[heights_cm, age_yrs],
    y=weights_kg,
    x_label=["Height (cm)", "Age (years)"],
    y_label="Weight (kg)",
    line_values=(surface_heights, surface_ages, surface_weights)
)
```

Although we cannot say anything about its quality at this stage, the plane can be used to predict weight from a given height and age.

For example, if a person is 175cm tall and 40 years old we can predict their weight:

```python
height = 175
age = 40
predicted_weight = linear_regression_multiple_inputs(height, age)

print("Using our linear regression we have:")
print(f"Height (Observed): {height}cm")
print(f"Age (Observed): {age} years")
print(f"Weight (Predicted): {predicted_weight}kg.")
```

This can also be visualised as:

```python
plot_scatter_plot(
    x=[heights_cm, age_yrs],
    y=weights_kg,
    x_label=["Height (cm)", "Age (years)"],
    y_label="Weight (kg)",
    line_values=(surface_heights, surface_ages, surface_weights),
    single_prediction=[height, age]
)
```
